#### Merg  use for joining dataframe

```python
df1=pd.DataFrame({
    'city':['karachi','Lahor','islamabad','rawalpindi'],
    'huminity':[44,39,34,34],
    'temperature':[46,34,30,32],

})
df1
```

```python
df2=pd.DataFrame({
    'city':['islamabad','karachi','kohat'],
    'huminity':[42,33,32]
})
df2
```

```python
#inner is deafult and use as intersection and outer use as a union and 
#same left is used leftjoin and right as a right join
df3=pd.merge(df1,df2,on='city')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='outer')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='left')
df3
```

```python
df3=pd.merge(df1,df2,on='city',how='right')
df3
```

```python
#which data com from which table
df3=pd.merge(df1,df2,on='city',how='outer',indicator=True)
df3
```

#### use of suffixes

```python
df1=pd.DataFrame({
    'city':['new york','chicago','islamabad'],
    'temperature':[44,39,34],
    'huminity':[40,35,30]
})
df1
```

```python
df2=pd.DataFrame({
    'city':['new york','chicago','washinton'],
    'temperature':[40,32,35],
    'huminity':[42,33,32]
})
df2
```

```python
df3=pd.merge(df1,df2,on='city',suffixes=('_left','_right'))
df3
```

### contengency table or crosstabs

```python
df=pd.read_excel('/home/attaullah/Documents/survey.xls')
df
```

```python
# showing frequency of handedness based on nationality 
#margin show the total
pd.crosstab(df.Nationality,df.Handedness,margins=True)
```

```python
# showing frequency of Sex based on nationality 
pd.crosstab(df.Nationality,df.Sex,margins=True)
```

```python
# showing frequency of handedness based on Age and so on...
pd.crosstab(df.Age,df.Handedness,margins=True)
```

```python
# put arry on both x axis and y axis
pd.crosstab(df.Handedness,[df.Name,df.Nationality,df.Sex],margins=True)
```

```python
pd.crosstab([df.Nationality,df.Sex],[df.Handedness],margins=True)
```

```python
#getting percentage at row level use normalize=index
pd.crosstab(df.Sex,df.Handedness,normalize='index')
```

```python
pd.crosstab(df.Sex,df.Handedness,margins=True)
```

```python
#find average age of male or female who are left handed
pd.crosstab(df.Sex,df.Handedness,values=df.Age,aggfunc=np.average)
```
